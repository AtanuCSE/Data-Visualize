# **MatplotLib Basics**

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('Price.csv')
print(df)
```

```
plt.figure(figsize=(8,8))
plt.plot(df['Month'], df['Item'], label='Item')
plt.show()
```

```
plt.figure(figsize=(8,8))
plt.plot(df['Month'], df['Item'], label='Item')
plt.plot(df['Month'], df['Money'], label='Paisa')
plt.legend()
plt.show()
```
