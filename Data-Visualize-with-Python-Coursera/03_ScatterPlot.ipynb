# **Seaborn Scatterplot and Regression Plot**

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('Price.csv', header=0)
```

```
sns.set(rc={'figure.figsize':(10,8)}, style='darkgrid')
sns.scatterplot(df['Month'], df['Money'], data=df)
plt.show()
```

Regression Line

```
sns.set(rc={'figure.figsize':(15,8)})
sns.regplot(df['Item'], df['Money'], data=df)
plt.show()
```
