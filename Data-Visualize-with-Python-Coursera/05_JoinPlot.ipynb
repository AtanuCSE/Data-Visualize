# **Seaborn JointPlot**

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('Price.csv', header=0)
print(df)
```

```
sns.jointplot(df['Item'], df['Money'], data=df)
```

```
sns.jointplot(df['Item'], df['Money'], data=df,kind='hex')
```

```
sns.jointplot(df['Item'], df['Money'], data=df, kind='reg')
```
