```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('PopulationEstimation.csv')
df
```

```
def plot_optional():
    import matplotlib as plt
    %matplotlib inline
    
    ax = df.plot(x='Country', y='pop', kind='scatter', 
                    c=['#e41a1c','#377eb8','#e41a1c','#4daf4a','#4daf4a','#377eb8','#4daf4a','#e41a1c',
                       '#4daf4a','#e41a1c','#4daf4a','#4daf4a','#e41a1c','#dede00','#ff7f00'], 
                    xticks=range(0,16), s=6*df['pop']/1000000,  alpha=.75, figsize=[20,10]);

    for i, txt in enumerate(df['Country']):
        print(str(i)+" "+str(txt))
        ax.annotate(txt, [df['Country'][i], df['pop'][i] ], ha='center')

plot_optional()
```
