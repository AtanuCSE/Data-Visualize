```
import matplotlib as mpl
mpl.get_backend()
import matplotlib.pyplot as plt
```

```
import numpy as np
import pandas as pd
```

# **Simple Scatter Plot**

```
data = {'Model1': [86,23,32,13,45,12],'Model2': [96,98,34,12,22,19], 'Model3': [56,23,44,12,32,33]}
df = pd.DataFrame (data, columns = ['Model1','Model2','Model3'], index=['I1', 'I2','I3','I4','I5','I6'])
df
```

```
x = df['Model1']
y = df['Model2']
plt.figure()
plt.scatter(x, y)
```

# **Different Colors and Point Size**

```
x = df['Model1']
y = df['Model2']
color = ['green', 'red', 'blue', 'yellow', 'black', 'orange']
plt.figure()
plt.scatter(x, y, s=100, c=color)
```

# **Plot Same Column but Different Colors**

```
x = df['Model1']
y = df['Model2']

plt.figure()
plt.scatter(x[:2], y[:2], s=100, c='red', label='First Two')
plt.scatter(x[2:], y[2:], s=20, c='blue', label='Last Values')
```

# **Add Label, Legend, and Title**

```
x = df['Model1']
y = df['Model2']

plt.figure()
plt.scatter(x[:2], y[:2], s=100, c='red', label='First Two')
plt.scatter(x[2:], y[2:], s=20, c='blue', label='Last Values')

plt.xlabel('The number of times the child kicked a ball')
plt.ylabel('The grade of the student')

plt.legend(loc=5, frameon=False, title='Legend')

plt.title('Relationship between ball kicking and grades')
```

# **Access Children Objects**

```
x = df['Model1']
y = df['Model2']

plt.figure()
plt.scatter(x[:2], y[:2], s=100, c='red', label='First Two')
plt.scatter(x[2:], y[2:], s=20, c='blue', label='Last Values')

plt.xlabel('The number of times the child kicked a ball')
plt.ylabel('The grade of the student')

plt.legend(loc=5, frameon=False, title='Legend')

plt.title('Relationship between ball kicking and grades')

plt.gca().get_children()
```

# **Use Children Objects**

```
from matplotlib.artist import Artist


x = df['Model1']
y = df['Model2']

plt.figure()
plt.scatter(x[:2], y[:2], s=100, c='red', label='First Two')
plt.scatter(x[2:], y[2:], s=20, c='blue', label='Last Values')

plt.xlabel('The number of times the child kicked a ball')
plt.ylabel('The grade of the student')

plt.legend(loc=5, frameon=False, title='Legend')

plt.title('Relationship between ball kicking and grades')


def rec_gc(art, depth=0):
    if isinstance(art, Artist):
        # increase the depth for pretty printing
        #print("  " * depth + str(art))
        for child in art.get_children():
            rec_gc(child, depth+2)

# Call this function on the legend artist to see what the legend is made up of
rec_gc(plt.legend())
```
